# Classification task on random number generators

## Importing necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split,GridSearchCV,cross_val_score
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier,GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

## Importing Dataset

```python
data = pd.read_csv("/kaggle/input/generator-data/TrainingData.csv")
data
```

```python
X = data['Bitstream'].apply(lambda x: [int(bit) for bit in x]).tolist()
X = np.array(X)
```

```python
y = data['class']
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

## Scaling Data

```python
ss = StandardScaler()
```

```python
x_tr_scl = ss.fit_transform(x_train)
x_ts_scl = ss.transform(x_test)
```

## Model Training and Evaluating

### Random Forest Classifier

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(x_tr_scl,y_train)
yp = rfc.predict(x_ts_scl)
accuracy_score(yp,y_test)
```

```python
knc = KNeighborsClassifier()
knc.fit(x_tr_scl, y_train)
yp = knc.predict(x_ts_scl)
accuracy_score(yp,y_test)
```

### Decision Tree Classifier

```python
dtc =DecisionTreeClassifier()

dtc.fit(x_tr_scl,y_train)
yp = dtc.predict(x_ts_scl)
accuracy_score(yp,y_test)
```

### Logistic Regression

```python
lgc = LogisticRegression()

lgc.fit(x_tr_scl,y_train)
yp = lgc.predict(x_ts_scl)
accuracy_score(yp,y_test)
```

### Ada Boosting Classifier

```python
adc = AdaBoostClassifier()

adc.fit(x_tr_scl,y_train)
yp = adc.predict(x_ts_scl)
accuracy_score(yp,y_test)
```

### K Nearest Neighbors

```python
knc = KNeighborsClassifier()

knc.fit(x_tr_scl,y_train)
yp = knc.predict(x_ts_scl)
accuracy_score(yp,y_test)
```

### Support Vector Classifier

```python
svc = SVC()

svc.fit(x_train,y_train)
yp = svc.predict(x_ts_scl)
accuracy_score(yp,y_test)
```

### Graident Boosting Classifier

```python
gbc = GradientBoostingClassifier()
gbc.fit(x_tr_scl,y_train)
yp = gbc.predict(x_ts_scl)
accuracy_score(yp,y_test)
```
